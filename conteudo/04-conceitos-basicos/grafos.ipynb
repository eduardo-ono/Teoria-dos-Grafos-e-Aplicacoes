## Resumo da Terminologia de Grafos

| Tipo | Arestas | Múltiplas Arestas | Loops |
| --- | --- | :-: | :-: |
| Grafo simples | Não-Orientadas | Não | Não
| Grafo Orientado Simples | Orientadas | Não | Não
| Multigrafo | Não-orientadas | Sim | Não
| Multigrafo Orientado | Orientadas | Sim | Sim
| Pseudografo | Não-orientadas | Sim | Sim
| Grafos Mistos | Orientadas e Não-Orientadas | Sim | Sim

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph() # Cria um gravo vazio
G.add_node('A')
G.add_node('B')
G.add_node('C')
G.add_node('D')
G.add_node('E')
nx.draw(G, node_size=600, node_color='coral', with_labels=True)
```

```python
G.add_edge('A','B')
G.add_edge('A','C')
G.add_edge('B','C')
G.add_edge('B','E')
G.add_edge('C','D')
G.add_edge('C','E')
nx.draw(G, node_size=600, node_color='coral', with_labels=True)
```

```python
options = {
    "font_size": 16,
    "node_size": 800,
    "node_color": 'coral',
    "with_labels": True,
}
nx.draw(G, **options)
```

```python
V = ['A','B','C','D','E'] # Lista dos vértices
E = [('A','B'), ('A','C'), ('B','D'), ('B','E'), ('D','E'), ('C','E')] # Lista das arestas

G.add_nodes_from(V)
G.add_edges_from(E)
nx.draw_circular(G, **options)
```

### Grau do Vértice

```python
print(G.degree())
print(nx.degree(G))
print(nx.degree(G)['A'])
```
