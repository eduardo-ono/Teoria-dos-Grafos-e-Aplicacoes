__DEFINIÇÃO (grafo planar).__ &nbsp;
Um grafo $G$ é dito __*planar*__ se puder ser representado graficamente sem que haja cruzamento de suas arestas.

---

### Exemplo

O grafo $G = K_4$ é planar, conforme a representação a seguir:

```python
import networkx as nx
import matplotlib.pyplot as plt

opt = {
    'node_size': 500,
    'node_color': 'coral',
    'edgecolors': 'black',
    'with_labels': True
}

E = [('A','B'), ('A','C'), ('A','D'), ('B','C'), ('B','D'), ('C','D')]
G = nx.Graph(E)
pos_circular = nx.circular_layout(G)
pos_planar = nx.planar_layout(G)
plt.figure(1, figsize=[9,3])

plt.subplot(1, 2, 1) # num_linhas, num_cols, index
nx.draw(G, pos_circular, **opt)

plt.subplot(1, 2, 2) # num_linhas, num_cols, index
nx.draw(G, pos=nx.planar_layout(G), **opt)
```
