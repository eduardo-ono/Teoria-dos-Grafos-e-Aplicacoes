## Grafos

$\qquad$ Em termos gerais, grafos são estruturas que representam relacionamentos par-a-par entre objetos. Os objetos, representados por pontos ou cículos, são chamados __*vértices*__ ou __*nós*__. As relações entre os objetos, representadas por linhas, são chamaas de __*arestas*__.


---

__DEFINIÇÃO (_grafo_).__&nbsp; Um __*grafo*__ $G$ pode ser definido como um par de conjuntos finitos $(V(G),E(G))$ e denotado por $G = (V,E)$, onde $V$ é um conjunto não vazio de __*vértices*__ (ou __*nós*__) e $E$ um conjunto de __*arestas*__, que pode ou não ser vazio, sendo cada aresta um conjunto de dois vértices de $V$.

---

&nbsp;

$\qquad$ Um grafo $G$ pode ser também denotado por $G = (V(G),E(G))$, onde $V = V(G)$ é o conjunto de vértices (ou nós) e $E = E(G)$ o conjunto de arestas (_edges_ em inglês).

$\qquad$ O conjunto de arestas $E$ pode ser representado por $E = \{e_1, e_2, ..., e_m\}$. Cada aresta $e_i \in E(G)$ também pode ser representado como um par não ordenado $(v_i, v_j)$, $1 \le i \le n$ e $i \le j \le n$.

$\qquad$ Seja $G = (V,E)$ onde $V = \{1,2,3,4,5\}$ e $E = \{\{1,2\}, \{1,3\}, \{1,5\}, \{2,3\}, \{3,4\}, \{4,5\}\}$ ou $E = \{(1,2), (1,3), (1,5), (2,3), (3,4), (4,5)\}$ como um conjunto de pares não ordenados.

$\qquad$ Embora o uso de números para denotar vértices ser perfeitamento aceito, o uso de letras pode facilitar a notação das aretas. Seja $G = (V,E)$ onde $V = \{A, B, C, D, E\}$ e $E = \{(A,B), (A,C), (A,E), (B,C), (B,D), (D,E)\}$. Nesse caso, $E$ também pode ser denotado como $E = \{AB, AC, AE, BC, BD\}$.


### Terminologias

* Se $e = uv$ é uma aresta de um grafo $G$, então:

    * $u$ e $v$ são vizinhos/adjacentes;
    * $u$ e $v$ são extremos da aresta $e$;
    * $e$ incide sobre os vértices $u$ e $v$. 

* Duas arestas são __*adjacentes*__ quando compartilham um mesmo vértice.

O __*grau de um vértice*__ $v \in V(G)$, denotado por $d_v$ ou $\deg v$, é o número de arestas que incidem sobre $v$: $$ d(v)=\sum_{j \in V} {w_{ij}} $$

### Exemplo

$\qquad$ Seja o grafo $G = (V,E)$ tendo como conjunto de vértices $V = \{A,B,C,D,E\}$ e o conjunto de arestas $E = \{AB, BC, CD, DE, EA\}$. Sua representação gráfica é dada por:

```python
%matplotlib inline

import networkx as nx
import matplotlib.pyplot as plt


nx_draw_opt = {
        'node_size': 800,
        'node_color': 'coral',
        'font_size': 16,
        'with_labels': True 
}

E = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('E', 'A')]
G = nx.Graph()
G.add_edges_from(E)
plt.figure(figsize=[5, 5])
nx.draw_circular(G, **nx_draw_opt)
```

## Tipos de Grafos

<hr>

__DEFINIÇÃO (subgrafo).__&nbsp; Um __*subgrafo*__ de um grafo $G = (V,E)$ é um subconjunto $G' = (V',E')$ tal que $V' \subseteq V$ e $E' \subseteq E$.

<hr>

* Por definição, o próprio grafo $G$ é subgrafo dele próprio.

<hr>

__DEFINIÇÃO (digrafo).__&nbsp;
Um __*grafo orientado*__, __*grafo dirigido*__ ou __*digrafo*__ é um grafo que possui um conjunto finito de arestas orientadas, também chamadas direcionadas ou dirigidas.

<hr>

### Exemplo

```python
E = [('A', 'B'), ('A', 'C'), ('B', 'C')]
G = nx.DiGraph(E)
plt.figure(figsize=[4, 4])
nx.draw(G, arrowsize=26, **nx_draw_opt)
```

<hr>

__DEFINIÇÃO (multigrafo).__ &nbsp;
Um __*multigrafo*__ é um grafo não orientado que possui no mínimo duas arestas paralelas.

<hr>

### Exemplo:

```python
G = nx.MultiDiGraph()
G.add_edges_from([
    ('A','B'),
    ('A','C'),
    ('B','C'),
    ('C','B'),
    ('C','A')
])
plt.figure(figsize=[5, 4])
nx.draw(G, arrowstyle='-', connectionstyle='arc3, rad=0.1', **nx_draw_opt)
```

```python
G = nx.MultiGraph ([(1,2),(1,2),(1,2),(3,1),(3,2)])
pos = nx.random_layout(G)
nx.draw_networkx_nodes(G, pos, node_color = 'coral', node_size = 600, alpha = 1)
ax = plt.gca()
for e in G.edges:
    ax.annotate("",
        xy = pos[e[0]], xycoords='data',
        xytext = pos[e[1]], textcoords='data',
        arrowprops = dict(arrowstyle="-", color="0.5",
            shrinkA=5, shrinkB=5, patchA=None, patchB=None,
            connectionstyle="arc3,rad=rrr".replace('rrr', str(0.3 * e[2])),
        ),
    )
plt.axis('off')
plt.show()
```

<hr>

Um __*grafo simples*__ é um grafo que não contém arestas paralelas e laços.

<hr>

<hr>

__DEFINIÇÃO (grafo completo).__&nbsp; Um grafo de $n$ vértices é dito __*completo*__ e denotado por $K_n$ quando há apenas uma única aresta unindo cada par de vértices.

<hr>

### Exemplos

```python
K1 = nx.complete_graph(4)
print(K1.nodes())
print(K1.edges())
print(K1.size())
plt.figure(figsize=[5, 5])
nx.draw(K1, **nx_draw_opt)
```

__TEOREMA (do Aperto de Mãos, Euler).__ &nbsp;
A soma dos graus dos vértices de um grafo (não orientado) $G$ é igual ao dobro do número de arestas de $G$, ou seja:

$$
\sum_{i=1}^{n}{d(v_i)} = 2 \cdot |E| = 2\,m
$$
