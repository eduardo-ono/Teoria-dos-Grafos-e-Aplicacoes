__DEFINIÇÃO.__ &nbsp;
Um grafo é dito __*bipartido*__ quando seus vértices podem ser divididos em dois grupos disjuntos tais que cada aresta ligue apenas vértices de grupos diferentes.

<hr>

```python
%matplotlib inline

import networkx as nx
from networkx.algorithms import bipartite

opt = {
    'node_size': 800,
    'node_color': 'coral',
    'font_size': 16,
    # 'edgecolors': 'black',
    'with_labels': True
}

G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4], bipartite=0)
G.add_nodes_from(['A', 'B', 'C'], bipartite=1)
G.add_edges_from([(1, 'A'), (1, 'B'), (2, 'B'), (2, 'C'), (3, 'C'), (4, 'A')])
top = nx.bipartite.sets(G)[0]
pos = nx.bipartite_layout(G, top)
nx.draw(G, pos, **opt)
print(nx.bipartite.sets(G))
```
