## Representação de um labirinto através de uma lista de adjacência

```python
import networkx as nx
import matplotlib.pyplot as plt

nx_draw_opt = {
    'node_size': 500,
    'node_color': 'coral',
    'font_size': 10,
    'with_labels': True
}

# Lista de adjacência
linhas = [
    'A1 A2 B1', 'A2 A3', 'A4 A5 B4', 'A5 A6 B5', 'A6 A7', 'A7 A8', 'A8 A9', 'A9 A10 B9', 'A10 B10',
    'B1 C1', 'B2 B3 C2', 'B3 B4', 'B4 C4', 'B5 B6', 'B6 B7 C6', 'B8 B9', 'B10 C10',
    'C1 C2 D1', 'C3 C4', 'C4 C5 D4', 'C5 D5', 'C6 C7', 'C7 D7', 'C8 C9 D8', 'C9 D9', 'C10 D10',
    'D2 D3', 'D3 D4 E3', 'D5 D6 E5', 'D6 D7 E6', 'D7 D8', 'D8 E8', 'D9 D10',
    'E1 E2 F1', 'E2 F2', 'E3 E4 F3', 'E4 E5', 'E6 E7', 'E7 F7', 'E8 E9 F8', 'E9 F9', 'E10 F10',
    'F1 G1', 'F2 G2', 'F3 F4', 'F4 G4', 'F5 G5', 'F6 F7 G6', 'F7 F8', 'F9 F10 G9', 'F10 G10',
    'G1 H1', 'G2 G3', 'G3 G4 H3', 'G4 G5', 'G5 G6 H5', 'G6 H6', 'G7 G8 H7', 'G8 G9 H8', 'G10 H10',
    'H1 H2', 'H2 I2', 'H3 I3', 'H4 H5 I4', 'H5 I5', 'H6 I6', 'H7 I7', 'H9 H10 I9',
    'I1 I2 J1', 'I2 I3 J2', 'I3 I4', 'I5 J5', 'I6 I7', 'I7 J7', 'I8 J8', 'I9 J9', 'I10 J10',
    'J1 J2', 'J3 J4', 'J4 J5', 'J5 J6', 'J6 J7', 'J7 J8', 'J9 J10'
]

G = nx.parse_adjlist(linhas)
nodes = [chr(letra) + str(i) for letra in range(ord('A'), ord('K')) for i in range(1, 11)]
G.add_nodes_from(nodes)
print('Vértices =', nodes)
# pos = nx.spring_layout(G, seed=5)
pos = nx.kamada_kawai_layout(G)
plt.figure(figsize=[14, 14])
nx.draw(G, pos, **nx_draw_opt)
```
