```python
%matplotlib inline

import networkx as nx
import matplotlib.pyplot as plt

nx_draw_opt = {
    'node_size': 800,
    'node_color': 'coral',
    'font_size': 16,
    'with_labels': True
}
```

```python
G = nx.complete_graph(5)
eulerianoOK = nx.is_eulerian(G)
print("Euleriano:", nx.is_eulerian(G))
nx.draw(G, **nx_draw_opt)
if eulerianoOK:
    print(list(nx.eulerian_circuit(G)))
    list(nx.eulerian_circuit(G, source=1))
    [u for u, v in nx.eulerian_circuit(G)]
```

```python
E = [('A', 'B'), ('A', 'C'), ('A', 'D'), ('A', 'E'),
     ('B', 'C'), ('B', 'G'),
     ('C', 'D'), ('C', 'E'),
     ('G', 'H'),
     ('H', 'I'), ('H', 'J'),
     ('I', 'J')
]
G = nx.Graph(E)
semiEulerianoOK = nx.is_semieulerian(G)
print("Semi Euleriano:", semiEulerianoOK)
nx.draw(G, **nx_draw_opt)
```
