## Algoritmo Não Recursivo

&nbsp;

```python
# O ALGORITMO A SEGUIR NÃO ESTÁ CORRETO!!!

def DFS(G, s, visitados):
    """Algoritmo: Busca em Profundidade (não-recursivo)"""
    visitados.append(s)
    pilha = [s]
    while len(pilha) > 0:
        v = pilha.pop()
        if v not in visitados:
            visitados.append(v)
        else:
            continue
        for vizinho in G[v]:
            pilha.append(vizinho)
```

### Exemplo

```python
%matplotlib inline

import networkx as nx
import matplotlib.pyplot as plt

nx_draw_opt = {
        'node_size': 800,
        'node_color': 'coral',
        'font_size': 16,
        'with_labels': True 
}
```

```python
E = [
    (1, 2), (1, 3),
    (2, 3), (2, 4), (2, 8), (2, 11),
    (3, 4), (3, 5),
    (4, 7), (4, 8), (4, 9),
    (5, 6),
    (7, 10), (7, 11),
    (9, 11),
    (10, 11)
]
G = nx.Graph()
G.add_edges_from(E)
nx.draw(G, **nx_draw_opt)
```

```python
# Busca em largura
visitados = []
DFS(G, 7, visitados)
print(visitados)
```
