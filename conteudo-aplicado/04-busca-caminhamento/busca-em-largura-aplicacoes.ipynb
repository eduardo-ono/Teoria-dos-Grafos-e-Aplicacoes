### Algoritmo

```python
import networkx as nx

def BFS(G, s, visitados):
    visitados.append(s)
    fila = [s]
    T = nx.Graph()
    while len(fila) > 0:
        v = fila.pop(0)
        T.add_node(v)
        for vizinho in G.neighbors(v):
            if vizinho not in visitados:
                visitados.append(vizinho)
                fila.append(vizinho)
                T.add_edge(v, vizinho)
    return T
```

```python
%matplotlib inline

import networkx as nx
import matplotlib.pyplot as plt

nx_draw_opt = {
        'node_size': 800,
        'node_color': 'coral',
        'font_size': 16,
        'with_labels': True 
}
```

```python
E = [
    (1, 2), (1, 3),
    (2, 3), (2, 4), (2, 8), (2, 11),
    (3, 4), (3, 5),
    (4, 7), (4, 8), (4, 9),
    (5, 6),
    (7, 10), (7, 11),
    (9, 11),
    (10, 11)
]
G = nx.Graph()
G.add_edges_from(E)
nx.draw(G, **nx_draw_opt)
```

```python
# Busca em largura
visitados = []
T = BFS(G, 7, visitados)
print(visitados)
print(T)
nx.draw(T, **nx_draw_opt)
```
