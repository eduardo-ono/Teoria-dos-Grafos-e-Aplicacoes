```python
import networkx as nx
import matplotlib.pyplot as plt

vertices_options = {
    'node_size': 600,
    'node_color': 'coral',
    'with_labels': True
}
```

```python
## Busca em Largura (BSF - Breadth-First Search)

def BFS(G, s): # G como lista de adjacências
    A = nx.to_dict_of_lists(G, nodelist=None)
    print(A)
    visitado = [0 for i in range(len(A) + 1)]
    fila = [s]
    visitado[s] = 1
    print(visitado)
    saida = [s]
    while len(fila) != 0:
        u = fila.pop(0)
        for v in A[u]:
            if visitado[v] == 0:
                fila.append(v)
                saida.append(v)
                visitado[v] = 1
    return saida
```

```python
E = [(1,2), (1,3), (2,3), (2,4), (2,8),
     (3,4), (3,5), (4,7), (4,8), (4,9),
     (5,6), (7,10), (7,11), (9,11), (10,11)]
G = nx.Graph()
G.add_edges_from(E)
nx.draw(G, **vertices_options)
lista_adjacencias = nx.generate_adjlist(G)
BFS(G, 7)
```
