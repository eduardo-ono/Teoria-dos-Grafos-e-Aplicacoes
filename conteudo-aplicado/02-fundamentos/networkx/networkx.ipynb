```python
options = {
        'node_size': 600,
        'node_color': 'coral',
        'with_labels': True
    }

G = nx.gnm_random_graph(8, 15)
nx.draw(G, **options)
```

```python
def nx_draw_opt(node_size=600, font_size=14, node_color='coral', with_labels=True):
    return {
        'node_size': node_size,
        'font_size': font_size,
        'node_color': node_color,
        'with_labels': with_labels
    }

G = nx.gnm_random_graph(8, 15)
nx.draw(G, **nx_draw_opt(font_size=16))
```

## "Inferindo" Grafos

### Descrição

$\qquad$ Dado um grafo já definido através da biblioteca NetworkX, podemos "inferir" (obter informações e resultados) utilizando métodos fornecidos pela mesma.

&nbsp;


| Método | Descrição |
| --- | --- |
| edges() | Retorna a lista de arestas do grafo.
| nodes() | Retorna a lista de vértices do grafo.
| num_of_nodes() | Retorna o número de vértices ($n$) do grafo.
| num_of_edges | Retorna o número de arestas ($m$) do grafo.
| size() | Retorna o tamanho (número de arestas) do grafo.

&nbsp;

| Método | Descrição |
| --- | --- |
| degree($nbunch=None$, $weight=None$) | Retorna o grau de um único vértice ou de $n$ vértices. Se $nbunch$ for omitido, retorna o grau de todos os vértices.

### Exemplos

```python
def nx_draw_opt(
        node_size=600,
        node_color='coral',
        with_labels=True
    ):
    return {
        'node_size': node_size,
        'node_color': node_color,
        'with_labels': with_labels
    }

G = nx.gnm_random_graph(8, 15)
nx.draw(G, **nx_draw_opt())
print('Vértices:', G.nodes())
print('Arestas', G.edges())
print('n =', G.number_of_nodes())
print('m =', G.number_of_edges())
print('Tamanho =', G.size())
print('Grau =', G.degree())
print('Grau dos vértices "0", "1" e "7" =', G.degree([0, 1, 7]))
print('Vértice "0" em V(G):', 0 in G.nodes())
print('Vértice "11" em V(G):', 0 in G.nodes())
print('Matriz de Adjacência:\n', nx.adjacency_matrix(G), sep='')
```
