## "Inferindo" Grafos

## Descrição

Dado um grafo já definido através da biblioteca NetworkX, podemos "inferir" (obter informações e resultados) utilizando métodos fornecidos pela mesma.

&nbsp;


| Método | Descrição |
| --- | --- |
| edges() | Retorna o
| nodes() | Retorna a lista de vértices do grafo.
| num_of_nodes() | 
| num_of_edges |
| size() | Retorna o número de arestas do grafo.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.gnm_random_graph(8, 15)
nx.draw(G, node_size=500, node_color='coral', with_labels=True)
print('Vértices:', G.nodes(), 'Tipo:', type(G.nodes))
print('Arestas', G.edges(), 'Tipo:', type(G.edges))
print('n =', G.number_of_nodes())
print('m =', G.number_of_edges())
```
