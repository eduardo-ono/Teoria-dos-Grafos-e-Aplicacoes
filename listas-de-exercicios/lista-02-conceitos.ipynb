### Exercício 1.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Considere os dois diagramas a seguir. Rotule os vértices e as arestas de modo que ambos os diagramas representem o mesmo grafo.

```python
import networkx as nx
import matplotlib.pyplot as plt

opt = {
    'node_size': 600,
    'node_color': 'coral'
}

E1 = [('A', 'B'), ('A', 'D'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('E', 'A')]
G1 = nx.Graph(E1)
E2 = [('A', 'C'), ('A', 'D'), ('B', 'D'), ('B', 'E'), ('C', 'D'), ('C', 'E')]
G2 = nx.Graph(E2)
plt.figure(1, figsize=(10, 5))
plt.subplot(1, 2, 1) # num_linhas, num_cols, index
nx.draw_circular(G1, **opt)
plt.subplot(1, 2, 2) # num_linhas, num_cols, index
nx.draw_spectral(G2, **opt)
```
