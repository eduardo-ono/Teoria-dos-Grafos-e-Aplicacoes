<h4>Exercício 1.</h4>
Considere os dois diagramas a seguir. Rotule os vértices e as arestas de modo que ambos os diagramas representem o mesmo grafo.

```python
import networkx as nx
import matplotlib.pyplot as plt

V1 = ['A','B','C','D','E']
E1 = [('A','B'), ('B','C'), ('C','D'), ('D','E'), ('E','A')]

V2 = ['A','B','C','D','E']
E2 = [('A','C'), ('A','D'), ('B','D'), ('B','E'), ('C','E')]

G1 = nx.Graph()
G1.add_nodes_from(V1)
G1.add_edges_from(E1)

G2 = nx.Graph()
G2.add_nodes_from(V2)
G2.add_edges_from(E2)

# pos = nx.spring_layout(Diagramas, iterations=100)

options = {
    'node_color': 'coral'
}
plt.figure(1, figsize=(10, 5))

plt.subplot(1, 2, 1) # num_linhas, num_cols, index
nx.draw_circular(G1)

plt.subplot(1, 2, 2) # num_linhas, num_cols, index
nx.draw_spectral(G2, font_size=8)
```
